*************************************************************************************************************************
#### summary(): 
            summary is a generic function used to produce result summaries of the results of various model fitting functions. The function invokes particular methods which depend on the class of the first argument

```r
summary(mtcars$mpg) # returns  'Min.''1st Qu.''Median''Mean''3rd Qu.''Max.' for the column mpg of mtcars dataset

#summary(mtcars)  

## ------------------------------------------------------------------------
s=summary(mtcars$mpg)

class(s) # returns type of output of summary for column "mpg" in dataset

names(s) # returns column names of table (of 'Min.''1st Qu.''Median''Mean''3rd Qu.''Max.') created by summary() function

s["1st Qu."]  # returns 1st Qu of column mpg from mtcars
s["Min."]  # returns min value of mpg column from mtcars dataset
```

```r
## ------------------------------------------------------------------------
summary(mtcars)  # use this to print table for all columns in dataset

s=summary(mtcars) # returns 'Min.''1st Qu.''Median''Mean''3rd Qu.''Max.' for each column in mtcars dataset

class(s)  
names(s)  # returns null as summary returns table with row as min median max and col with its values for each column

# to print mean mediam min max etc values of each column you can use it using slicing
s[1:6]
```

```r
# syntax: 
# s[rows numbers, column numbers i.e. name of columns]
s[1,1]  # returns first row (Min) of first colunmn (mpg) 

#QUE. get mean value of wt (6) column
#ANS: index(row) of means is 4 and index of col wt is 6 so passing 4,6 to get its mean
print("----------------------------------------------------------------------")
print("mean value of wt (6) column is: ")
s[4,6]
print("----------------------------------------------------------------------")
#QUE:  get MIN of all columns
#ANS: to get MIN of all columns in dataset need to pass index of min(1) and all index of columns(1 to 11)
print("MIN of all columns: ")
s[1,1:11]
print("----------------------------------------------------------------------")
#QUE: get all function values for "disp" column 
#ANS: pass all indexs of rows (1:6) and index of disp column (3)
print("All Function values of 'disp' column: ")
s[1:6, 3]

# like above you can perform different operations using slicing methods and data
```

<h3>str():</h3>
        <i>Display the internal Structure of an Object.</i>
        <br>str() function in R Language is used for compactly displaying the <b>internal structure</b> of a R object. <u>It can display even the internal structure of large lists which are nested</u>. It provides one liner output for the basic R objects letting the user know about the object and its constituents. It can be used as an alternative to <b>summary()</b> but <b>str()</b> is more compact than <b>summary()</b>. It gives information about the <b>rows(observations)</b> and <b>columns(variables)</b> along with additional information like the names of the columns, class of each columns followed by few of the initial <b>observations (rows)</b> of each of the columns.

```r
# Here we created one temp list for understanding
# str() function returns compact summary of each value in list along with its data type and few 
# row values (here it returns only one value as list contains only one value for each column.)
# we passed list of 10 values so str() function considered 10 columns for list and all 10 values as single row(observation)

# we passed list(4,5,4,2) so it also display the internal structure of lists which are nested.

tempList <- list(2, list(4,5,4,2), 6, 7, 9, 13i, "String data", 3.5, 1L)
str(tempList)
```

```r
# str(mtcars) will return all compact internal structure with its all column names with 
# its data type and also shows some rows of data from dataset. 
str(mtcars)
```

<h3> Histogram:</h3>
A histogram contains a rectangular area to display the statistical information which is proportional to the frequency of a variable and its width in successive numerical intervals. A graphical representation that manages a group of data points into <b>different specified ranges</b>. <em>It has a special feature which shows no gaps between the bars and is similar to a vertical bar graph.

<b>Syntax:</b> 
<br>             hist(v, main, xlab, xlim, ylim, breaks, col, border)

<b>Parameters: </b>
<br><u>v:</u>        This parameter contains numerical values used in histogram.
<br><u>main: </u>      This parameter main is the title of the chart.
<br><u>col: </u>       This parameter is used to set color of the bars.
<br><u>xlab:</u>       This parameter is the label for horizontal axis.
<br><u>border:</u>   This parameter is used to set border color of each bar.
<br><u>xlim:</u>   This parameter is used for plotting values of x-axis.
<br><u>ylim:</u>   This parameter is used for plotting values of y-axis.
<br><u>breaks:</u>   This parameter is used as width of each bar.</em>

https://www.geeksforgeeks.org/histograms-in-r-language/

```r
# sprintf() is a c library function used here to print values here. no need to focus much
sprintf("total rows for mpg: %s ",length(mtcars$mpg))

# to count number of values present in list between two values we can use sum() function, multiple solutions can be used.
# here we counted number of values between 10 and 15 for mpg column in mtcars dataset, same applied for other ranges
# based on count (i.e. frequency in below histogram) histogram created. 
sum(mtcars$mpg >10 & mtcars$mpg <=15)
sum(mtcars$mpg >15 & mtcars$mpg <=20)
sum(mtcars$mpg >20 & mtcars$mpg <=25)
sum(mtcars$mpg >25 & mtcars$mpg <=30)
sum(mtcars$mpg >=30 & mtcars$mpg <=35)

# plots histgram for mpg column of mtcars dataset, range is given on X axis and its frequency (count) is on Y axis
hist(mtcars$mpg)
```
